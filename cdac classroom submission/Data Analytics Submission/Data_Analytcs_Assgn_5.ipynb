```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df.drop(columns=('Species')))
plt.title('boxplot of numerical columns')
```

```python
df = df[(df['SepalWidthCm'] >=2.2 )& (df['SepalWidthCm'] <=4.0)]
```

```python
plt.figure(figsize=(10,8))
sns.boxplot(data=df.drop(columns=('Species')))
plt.title('boxplot of numerical columns')
plt.show()
```

```python
#1. Comparison Between Various Species Based on Sepal Length and Sepal Width
plt.figure(figsize=(10,8))
sns.scatterplot(x='SepalWidthCm',y='SepalLengthCm',hue='Species',palette='Set2',data=df)
plt.title('Comparison on sepal length and sepal width')
plt.xlabel('SepalWidth')
plt.ylabel('SepalLength')
plt.legend(title='Species')
plt.show()
```

```python
#2. Bi-variate Analysis via Pairplot Visualization
sns.pairplot(data=df,hue='Species',palette='Set2')
plt.show()
```

```python
#3. Correlation Between Sepal Length and Sepal Width
plt.figure(figsize=(8, 5))
sns.regplot(x='SepalLengthCm', y='SepalWidthCm', data=df, scatter_kws={'s': 50}, color='green')
plt.title('Scatter Plot with Regression Line for Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.show()
```

```python
#plt.figure(figsize=(8, 5))
plt.figure(figsize=(16, 10))

for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']):
    plt.subplot(2, 2, i+1)
    sns.violinplot(x='Species', y=feature, data=df, palette='Set2')
    plt.title(f'Violin Plot of {feature} by Species')

plt.tight_layout()
plt.show()
```

```python
#5. Plot the Density of Length and Width in Each Species
plt.figure(figsize=(16, 10))

for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']):
    plt.subplot(2, 2, i+1)
    sns.kdeplot(data=df, x=feature, hue='Species', fill=True, common_norm=False, palette='Set2')
    plt.title(f'Density Plot of {feature} by Species')

plt.tight_layout()
plt.show()
```

```python
#6. Plot the Probability Density Function (PDF) with Histograms and KDE Plots
plt.figure(figsize=(16, 10))

for i, feature in enumerate(['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']):
    plt.subplot(2, 2, i+1)
    sns.histplot(data=df, x=feature, kde=True, hue='Species', element='step', palette='Set2')
    plt.title(f'Histogram and KDE for {feature}')
    plt.xlabel(feature)
    plt.ylabel('Density')

plt.tight_layout()
plt.show()
```
