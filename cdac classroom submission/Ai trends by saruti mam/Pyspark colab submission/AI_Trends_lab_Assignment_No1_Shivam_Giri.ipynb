```python
mt = spark.read.csv("mtcars.csv",header=True)
```

```python
mt.show(10)
```

```python
from pyspark.sql.functions import *
mt.orderBy(desc("mpg")).show(5)
```

```python
mt.filter(mt.mpg < 18).show()
```

```python
from pyspark.sql.types import *
mt = mt.withColumn("wtTon",col("wt")*0.45)
```

```python
mt.show()
```

```python
mt = mt.withColumn("wt",col("wt").cast("Integer"))
```

```python
mt.groupBy("cyl").avg("wt").show()
```

```python
mt.createOrReplaceTempView("mt1")
```

```python
spark.sql("select distinct gear from mt1 where cyl between 4 and 9").show()
```
